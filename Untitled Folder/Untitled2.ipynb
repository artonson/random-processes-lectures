```python
timestamps = np.linspace(0.01, 10, 1000)
dt = 1e-2
x_0 = 2.
sigma = 0.1
mu = 1.2
theta = 1.
dbm = np.random.normal(size=1000)
ou = np.zeros_like(dbm)
ou[0] = x_0
for i in range(1, len(dbm)):
    ou[i] = ou[i - 1] + theta * (mu - ou[i - 1]) * dt + sigma * dbm[i] * np.sqrt(dt)
```

```python
pgf_with_pdflatex = {
    "text.usetex": "False",
    "font.size": "12",
    "font.family": "serif"
}
matplotlib.rcParams.update(pgf_with_pdflatex)
```

```python
fig = figure(figsize=(12, 4))
subplot(1, 2, 1)
plot(timestamps, dbm)
title('Смещения броуновского движения')
subplot(1, 2, 2)
plot(timestamps, ou)
title('Траектория процесса Орнштейна-Уленбека')
```

```python
fig.tight_layout()
fig.savefig('ornstein-uhlenbeck-integration.eps', format='eps')
```
