```python
%pylab inline
def compute_ou_cov(timestamps):
    n = len(timestamps)
    R = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            R[i, j] = np.exp(-np.abs(timestamps[i] - timestamps[j]))
    return R

timestamps = np.linspace(0.01, 10, 1000)

cov_ou = compute_ou_cov(timestamps)
cov_ou_chol = np.linalg.cholesky(cov_ou)

pgf_with_pdflatex = {
    "text.usetex": "False",
    "font.size": "12",
    "font.family": "serif"
}
matplotlib.rcParams.update(pgf_with_pdflatex)
fig = figure(figsize=(11, 4))
subplot(1, 2, 1)
title(r'Матрица ковариаций')
imshow(cov_ou, interpolation='nearest')
subplot(1, 2, 2)
title(r'Ковариационная функция $R(5, t)$')
plot(timestamps, cov_ou[500, :])
fig.tight_layout()
fig.savefig('ornstein-uhlenbeck-covariation.eps', format='eps')
```
