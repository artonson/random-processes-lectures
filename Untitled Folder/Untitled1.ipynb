```python
timestamps = timestamps = np.linspace(1e-2, 100, 100000)

dbm = np.random.normal(size=100000)
bm = 1e-1 * np.cumsum(dbm)

pgf_with_pdflatex = {
    "text.usetex": "False",
    "font.size": "12",
    "font.family": "serif"
}
matplotlib.rcParams.update(pgf_with_pdflatex)
```

```python
fig = figure(figsize=(12, 4))
subplot(1, 2, 1)
plot(timestamps, dbm)
title('Смещение броуновского движения')
subplot(1, 2, 2)
plot(timestamps, bm)
title('Траектория броуновского движения')
```

```python
fig.tight_layout()
fig.savefig('brownian-motion-integration.eps', format='eps')
```
